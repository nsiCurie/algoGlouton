## Recherche d'une occurrence d'un tableau
On dispose maintenant d’un tableau tableau1 de 101 éléments mélangés.<br>
On veut chercher le rang  d’une occurrence dans ce tableau.<br>
4)	Ecrire une fonction recherche qui prend en paramètre le tableau et une valeur, et qui renvoie le rang de la valeur.<br>
5)	Ecrire l’algorithme d’une fonction paire prenant en argument un tableau et qui renvoie un tableau  contenant tous les rangs des occurrences paires de l’argument.<br>
6)	Ecrire un algorithme d’une fonction moyenne prenant en argument un tableau et qui renvoie la moyenne de  toutes les occurrences paires du tableau.<br>

```python
def recherche(tableau,valeur):
    i=0
    while tableau[i]!=valeur:
        i=i+1
    return i

print(recherche(tableau1,10))
```

```python
def paire(tableau):
    tableau3=[]
    for i in range(len(tableau)):
        if tableau[i]%2==0:
            tableau3.append(i)
    return tableau3

print(paire(tableau1))
```

```python
def moyenne(tableau):
    occurrence=paire(tableau)
    somme=0
    for i in occurrence:
        somme=somme+tableau[i]
    return somme/len(occurrence)

print(moyenne(tableau1))
```

## IV. Un tableau plus complexe et vers une modélisation d’un échantillon radioactif.
7)	Ecrire un tableau en compréhension qui simule 10000 lancers de dé à 6 faces.


```python
lancer=[random.randint(1,6) for x in range(10000)]
print(lancer)
```

8)	Ecrire l’algorithme d’une fonction qui prend en argument un lancer et qui renvoie un tableau contenant les nombres de lancers de chaque valeur.<br>
(Exemple : [18,15,14,11,5,3] indique qu’il y a eu 18 fois le 1, 15 fois le 2 etc… pour 66 lancers au total)


```python
def compte(tableau):
    resultat=[0,0,0,0,0,0]
    for i in tableau:
        resultat[i-1]+=1
    return resultat

print(compte(lancer))
```

9)	Ecrire un algorithme qui recherche dans le lancer le chiffre tiré  le maximum de fois.

```python
def chiffreMax(tableau):
    max=0
    for i in range(len(tableau)):
        if tableau[i]>tableau[max]:
            max=i
    return max+1
resultat=compte(lancer)
print(resultat)
print(chiffreMax(resultat))
```

10)	Lors d’une désintégration radioactive, un noyau instable a une probabilité p de se désintégrer : p=1/6.</p>
La fonction ci-dessous renvoie le nombre de noyaux restants si n noyaux sont présents dans un échantillon radioactifs et que la probabilité de désintégration  est p. 
Tester cette fonction avec 10000 noyaux et p=1/6.


```python
def compte_noyaux_restants(n,p):
    N=n
    for i in range(n):
        if random.random()<p:
            N=N-1
    return N

print(compte_noyaux_restants(10000,1/6))
```

11)	Ecrire un algorithme qui renvoie  la liste Liste_nb_noyaux_restants contenant le nombre de noyaux restants à chaque étape jusqu’à ce qu’il n’en reste plus que 1 % du nombre initial. On utilisera la fonction précédente.

```python
def desintegration(n,p):
    Liste_nb_noyaux_restants=[]
    N=n
    while N>n*0.01:
        N=compte_noyaux_restants(N,p)
        Liste_nb_noyaux_restants.append(N)
    return  Liste_nb_noyaux_restants


Liste_nb_noyaux_restants = desintegration(10000,1/6)

print(Liste_nb_noyaux_restants)
```

12)	Ecrire un algorithme qui permet de déterminer la demi-vie de la substance radioactive (nombre d’étapes successives pour que le nombre de noyaux soit égal  à la moitié des noyaux initiaux).	

```python
def demiVie(n,liste):
    for i in range(len(liste)):
        if liste[i]<n/2:
            return i+1

print(demiVie(10000,Liste_nb_noyaux_restants))
```
